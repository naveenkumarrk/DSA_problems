1- Given an array of 1s and 0s which has all 1s first followed by all 0s. Find the number of 0s.

 Count the number of zeroes in the given array. Please try to solve it in O(log(n)).

```python
nums = [1,1,1,1,0,0]

def first_zero(nums):
    n = len(nums)
    l = 0
    r = n-1
    while l<= r:
        mid = l + (r-l)//2
        if nums[mid] == 0:
            r = mid -1
        else:
            l = mid+1
    if l < n and nums[l] == 0:
        return n- l
    else:
        return 0
    
first_zero(nums)
```

3-  Reverse a Queue using Recursion.

```python
from collections import deque

nums = deque()
nums.append(4)
nums.append(134)
nums.append(45)
nums.append(8)
nums.append(20)


def reverse_q(nums):
    n = len(nums)
    if n ==1:
        return nums[n]
    return reverse_q(nums)
reverse_q(nums)
```
